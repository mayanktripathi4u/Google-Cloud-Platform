## Step 1: Create Table Reference

```python
dataset_ref = bigquery.DatasetReference(client.project, 'streamingDataSet')
table_ref = bigquery.TableReference(dataset_ref=dataset_ref, table_id='tbl_streamingData')
```

## Step 2.1: Create labels

```python
labels = {
    'type': 'social_media',
    'category' : 'cloud_computing'
}
```

```python
myTbl = client.get_table(table_ref)
myTbl
```

```python
myTbl.labels = labels
```

```python
myTbl = client.update_table(myTbl, ["labels"])
```

## Step 2.2: Add label to the table reference.

```python
new_labels = {
    'type' : 'software',
    'color': 'green',
    'year': '2024'
}

myTbl = client.get_table(table_ref)
myTbl.labels = new_labels
myTbl = client.update_table(myTbl, ["labels"])
```
